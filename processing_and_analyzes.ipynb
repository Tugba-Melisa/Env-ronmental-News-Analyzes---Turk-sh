```python
from gensim import corpora, models, similarities 
import gensim
```

```python
tokenized = data['content-token']
dictionary = corpora.Dictionary(tokenized)
dictionary.filter_extremes(no_below=1, no_above=0.8)
corpus = [dictionary.doc2bow(tokens) for tokens in tokenized]
```

```python
from gensim.models.coherencemodel import CoherenceModel
num_topics_range = [6, 8, 10, 12, 15]
passes_range = [1, 5]
iterations_range = [50, 100]
best_params = {}
for num_topics in num_topics_range:
    for passes in passes_range:
        for iterations in iterations_range:
            # LDA modelini eğit
            lda_model = models.LdaModel(corpus, num_topics=num_topics, 
                                        id2word=dictionary, passes=passes,
                                        iterations=iterations) 
            # Cohere1nce skorunu hesapla
            coherence_model = CoherenceModel(model=lda_model, corpus=corpus,
                                             dictionary=dictionary, coherence='u_mass')
            score = coherence_model.get_coherence()
            print(score)
            params = {'num_topics': num_topics, 'passes': passes, 'iterations': iterations}
            print(params)
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics =50, 
                                           id2word=dictionary, passes=5, 
                                           iterations=500)
ldamodel.save('mOdel.gensim')
topics = ldamodel.print_topics(num_words=30)
```

```python
def find_max_topic_for_document(doc, ldamodel):
    get_document_topics = ldamodel.get_document_topics(doc,  minimum_probability=0.0)
    second_elements = [item[1] for item in get_document_topics]
    max_value = max(second_elements)
    num_topic = second_elements.index(max_value)
    return num_topic

# Her doküman satırı için num_topic'leri bulun
data['num_topic'] = [find_max_topic_for_document(doc, ldamodel) for doc in corpus]
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


def extract_keywords(sentence, keywords):
    return [keyword for keyword in keywords if keyword in sentence]

# Count the occurrence of each keyword in the dataset
keyword_counts = {}
for sentence in data['content']:
    keywords_found = extract_keywords(sentence.lower(), environmental_key)
    for keyword in keywords_found:
        if keyword in keyword_counts:
            keyword_counts[keyword] += 1
        else:
            keyword_counts[keyword] = 1

# Create DataFrame from keyword counts
keyword_df = pd.DataFrame.from_dict(keyword_counts, orient='index', columns=['count'])
keyword_df.index.name = 'keyword'
keyword_df.sort_values(by='count', ascending=False, inplace=True)

# Plot keyword counts
plt.figure(figsize=(12, 6))
sns.barplot(x=keyword_df.index[:20], y=keyword_df['count'][:20])
plt.title('Top 20 Keywords Frequency')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Keyword')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Initialize an empty Counter to store keyword counts for each source
source_keyword_counts = {source: Counter() for source in df['Source'].unique()}

# Iterate through each row of the DataFrame
for index, row in df.iterrows():
    keywords = row['Keywords']
    source = row['Source']
    # Count occurrences of each keyword for the current source
    source_keyword_counts[source].update(keywords)

# Select top 50 most common keywords across all sources
top_keywords = Counter()
for source_counts in source_keyword_counts.values():
    top_keywords.update(source_counts)
top_keywords = dict(top_keywords.most_common(20))

# Filter keyword counts for top 50 keywords
for source, counts in source_keyword_counts.items():
    source_keyword_counts[source] = {keyword: count for keyword, count in counts.items() if keyword in top_keywords}

# Create a DataFrame from the Counter
keyword_source_counts = pd.DataFrame(source_keyword_counts).fillna(0)

# Plotting
plt.figure(figsize=(12, 8))
keyword_source_counts.plot(kind='bar', stacked=True)
plt.title('Top 50 Keyword Distribution by Source')
plt.xlabel('Keywords')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Source', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from nltk.stem import PorterStemmer

# Sample DataFrame

# Initialize a PorterStemmer object for stemming
stemmer = PorterStemmer()

# Initialize an empty Counter to store combined keyword counts for each source
source_keyword_counts = Counter()

# Iterate through each row of the DataFrame
for index, row in df.iterrows():
    keywords = row['Keywords']
    source = row['Source']
    # Stemming each keyword and combine occurrences of similar keywords
    stemmed_keywords = tuple(sorted(set([stemmer.stem(keyword) for keyword in keywords])))  
    # Count occurrences of combined keywords for the current source
    source_keyword_counts[stemmed_keywords] += 1

# Convert the Counter into a DataFrame
source_keyword_df = pd.DataFrame.from_dict(source_keyword_counts, orient='index', columns=['Keyword Count'])

# Plotting
plt.figure(figsize=(12, 6))
source_keyword_df.sort_values(by='Keyword Count', ascending=False).plot(kind='bar', ax=plt.gca())
plt.title('Keyword Contribution by Source (Combined with Stemming)')
plt.xlabel('Combined Keywords')
plt.ylabel('Keyword Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
